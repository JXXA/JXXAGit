```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## matplot을 이용한 그래프 생성

```python
# 그래프를 그릴 데이터 생성
x = np.linspace(0, 10 ,11)
y = x ** 2 + x + 2 + np.random.randn(11)

print(x)
print(y)
```

### 1) 그래프 기본 생성

```python
plt.plot(x, y)
```

```python
## 주석 추가
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('X-Y Test Plot')
plt.grid(True)

## x,y축 범위 지정
plt.xlim(0, 20)
plt.ylim(0, 200)

plt.plot(x, y)
```

### 2) Parameter를 통한 그래프 형태 제어

```python
# 색상 변경
plt.plot(x, y, '#ff00ff')
```

```python
## 선 스타일 변경
plt.plot(x, y, 'r-.')
```

```python
plt.plot(x, y, 'p:')
```

```python
plt.plot(x, y, 'g^')
```

```python
## 라인 두께 변경
plt.plot(x, y, 'r-.', linewidth=9)
```

### 3) Subplot으로 여러 그래프 출력

```python
# 여러가지 그래프를 구역으로 나누어 그래프 그리기
plt.subplot(2, 2, 1)
plt.plot(x,y,'r')

plt.subplot(2, 2, 2)
plt.plot(x,y,'g')

plt.subplot(2, 2, 3)
plt.plot(y,x,'k')

plt.subplot(2, 2, 4)
plt.plot(y,np.exp(x),'b')
```

### 4) Historgarm 생성

```python
data = np.random.randint(1, 100, size = 200)
```

```python
plt.hist(data, bins=25, alpha=0.3)
plt.xlabel('값')
plt.ylabel('개수')
plt.grid(True)
```
