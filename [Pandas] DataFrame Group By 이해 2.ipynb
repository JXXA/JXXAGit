## 1. Pivot과 Pivot_table 함수

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })
df
```

#### 1. pivot 함수
* Index 값에는 중복 데이터가 사용될 수 없다. 
    * 'ValueError: Index contains duplicate entries, cannot reshape' 오류 발생

```python
df.pivot('지역', '요일')
```

```python
df.pivot('요일', '지역', '강수량')
```

#### 2. Pivot_table
* pivot과 동일하나, 중복값 존재 시 aggregation 함수 사용하여 값 채움

```python
pd.pivot_table(df, index='요일', columns='지역', aggfunc=np.mean)ㅠ
```

## 2. Stack과 Unstack
* Stack : 컬럼 레벨 -> 인덱스 레벨로 Dataframe 변경
    * 데이터를 쌓아올리는 개념
* Unstack : 인덱스 레벨 -> 컬럼 레벨로 Dataframe 변경

```python
new_df = df.set_index(['지역','요일'])
new_df
```

```python
new_df.unstack(0)
```

```python
new_df.unstack(1)
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```

## 3. Concat을 이용한 데이터 프레임 병합

#### 1. 컬럼명이 같은 Dataframe 병합

```python
df1 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
df2 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
```

```python
df2
```

```python
## 행 레벨 단위 병합
# Index를 무시할 경우 ignore_index 옵션 사용
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

#### 2. 컬럼명이 다른 Dataframe 병합

```python
df3 = pd.DataFrame({'key2' : np.arange(10), 'value2' : np.random.randn(10)})
```

```python
pd.concat([df1, df3])
```

## 4. Merge와 Join을 이용한 데이터 프레임 병합

#### 1. Merge
* SQL의 Join처럼 특정 컬럼을 기준으로 병합
    * Join 방식은 how 파라미터를 통해 명시
* pandas.merge 함수 사용

```python
customer = pd.DataFrame({'customer_id' : np.arange(6), 
                    'name' : ['철수'"", '영희', '길동', '영수', '수민', '동건'], 
                    '나이' : [40, 20, 21, 30, 31, 18]})

customer
```

```python
orders = pd.DataFrame({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9], 
                    'item' : ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'], 
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
orders.head()
```

```python
pd.merge(customer, orders, on='customer_id')
```

```python
pd.merge(customer, orders, on='customer_id', how='left')
```

```python
cust1 = customer.set_index('customer_id')
order1 = orders.set_index('customer_id')
```

```python
## Index를 기준으로 사용했기때문에 on을 표기할 필요가 없다.
pd.merge(cust1, order1, left_index=True, right_index=True)
```

#### ※ 응용 실습

```python
## 1. 가장 많이 팔린 아이템은?
pd.merge(customer, orders, on='customer_id').groupby('item').sum().sort_values(by='quantity', ascending=False)
```

```python
## 영희가 많이 구한 아이템은?
pd.merge(customer, orders, on='customer_id').groupby(['name','item']).sum()
```

#### 2. Join 함수

```python
## 인덱스가 있는 경우, 손 쉬운 병합을 위해 Join 함수
cust1.join(order1, how='inner')
```
