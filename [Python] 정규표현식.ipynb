## 02. 정규표현식 패턴

### 1) Module 지원 기능
* **re.search('찾을값','대상')** : 찾고자 하는 값에 대한 패턴 검색 수행
* start() : 패턴의 시작 Index 반환
* end() : 패턴의 종료 index 반환
* group() : 패턴 값

```python
import re
```

```python
m = re.search(r'search','search_value')
# <re.Match object; span=(0, 6), match='search'>

print(m.start()) # 시작위치
print(m.end()) # 종료위치
print(m.group()) ## 매칭되는 그룹 값

b = re.search(r'\d\d\w', '12dkgksdg11')
b.start()
b.end()
b.group()

c = re.search(r'..\w\w', '@$ABC')
c
```

### 2) [패턴] 범위
* [] 내부의 **메타 캐릭터는 단어 자체**를 나타냄
    * [abc.^] : a or b or c or . or ^ (= 특수한 문자가 와도 문자 그대로를 나타냄)
* [a-d] : a와 d 사이에 속하는 문자
* [0-9], [A-Z]. [a-z], [A-Za-z]
* [^0-9] : 0부터 9까지 속하는 숫자가 아닌 것

```python
re.search(r'[cbm]ct', 'cct')

re.search(r'[^abc]aron', 'faron')
```

### 3) [패턴] 특수문자


```python
re.search('\sand', 'apple and banana')
```

```python
re.search(r'\.and', '.and')
```

```python
re.search(r'p.g', 'pig')
```

### 3) [패턴] 반복되는 패턴


```python
re.search(r'a[bcd]*b', 'abllllb')

re.search(r'b\w+a', 'banana')

re.search(r'i+','piigiii')

## +와 *의 차이
re.search(r'pi+g','pg') # 한번이상은 나와야하기때문에 None 반환됨
re.search(r'pi*g','pg')

# 0개 또는 1개 문자 표현이기에 둘 다 결과가 나옴
re.search(r'https?','https://www.naver.com') 
re.search(r'http?','https://www.naver.com')
```

```python
re.search(r'b\w+a', 'cabana')
# <re.Match object; span=(2, 6), match='bana'>

# 문자열의 시작과 끝을 명시
re.search(r'^b\w+a', 'banana')
re.search(r'b\w+a$', 'pbanaa')
```

### 4) grouping
* \(\)을 사용하여 grouping
* 결과를 각 그룹별로 반환 가능
* 패턴 명시 할 경우, 각 그룹을 괄호 안에 넣어 분리하여 사용

```python
m = re.search(r'([a-zA-Z]+)@([a-zA-Z]+.[a-zA-Z]+)','test@test.com')
print(m.group(0))
print(m.group(1))
print(m.group(2))
```
