#### 3) [NaN 처리] 데이터 치환
* **fillna 함수** 사용

```python
# Age 변수의 NaN 값을 Age 평균으로 대체
train_data['Age'].fillna(train_data['Age'].mean())
```

```python
##################################################################
# 생존자/사망자 나이의 NaN 값을 평균으로 대체
##################################################################

## 생존자 나이 평균 구하기 
mean0 = train_data[train_data['Survived'] == 1]['Age'].mean()

## 사망자 나이 평균 구하기
mean1 = train_data[train_data['Survived'] == 0]['Age'].mean()

print(mean0, mean1)
```

```python
train_data[train_data['Survived'] == 1]['Age'].fillna(mean0) # 생존자 나이 평균 치환
train_data[train_data['Survived'] == 0]['Age'].fillna(mean1) # 사망자 나이 평균 치환
```

```python
# 실제 데이터에 치환 적용 (loc 함수 사용하여 Survived가 1인 Age 값 가져오기)
train_data.loc[train_data['Survived'] == 1,'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(mean0)
```

## 5. 범주형 데이터 - 타입 변환

```python
#train_data.info()
train_data.head()
```

#### 1) [Pclass 변수 변환] asype 를 이용한 타입 변환

```python
train_data['Pclass'] = train_data['Pclass'].astype(str)
```

#### 2) [Age 변수 변환] apply 를 이용한 타입 변환
* 변환 로직을 함수로 만들어, apply 함수 적용

```python
def age_transfer(age):
    if math.isnan(age):
        return -1
    
    return math.floor(age / 10) * 10
```

```python
train_data['Age'].apply(age_transfer) 
```

## 6. 범주형 데이터 - 전처리 (One-hot Encoding)
* 범주형 데이터의 분석을 위해서는 **숫자형 변환**이 필요

#### One-hot encoding
* 범주형 데이터의 **각 범주를 Column 레벨로 변경**
* 해당 범주에 해당 시 1, 아니면 0으로 채우는 인코딩 기법
* pandas.get_dummies 함수 사용

```python
# one-hot enconding 수행 시, 해당 범주 범위만큼의 컬럼 생성
pd.get_dummies(train_data, columns=['Pclass', 'Sex', 'Embarked'])

# 아래 데이터 중, Pclass의 경우 각 값인 1, 2, 3에 대한 컬럼이 생성 됨 확인
# Pclass_1 Pclass_2 Pclass_3
```

```python
# 범주 레벨로 올렸을 때, 첫번째 카테고리는 사용하지 않음 (일반적으로 True로 함)
pd.get_dummies(train_data, columns=['Pclass', 'Sex', 'Embarked'], drop_first=True)

########## drop_first 해석 ############

# 아래 데이터와 같이 pclass의 경우, Pcalss_1, Pcalss_2, Pcalss_3 컬럼 생성됨
# 이때 drop_first 줄 경우, Pcalss_1 가 사라짐

# 그럼 이때, 사라진 Pcalss_1 에 해당하는걸 찾으려면?
# Pclass_2, Pclass_3이 둘다 0일 경우 Pcalss_1에 해당하게 됨!
# 이렇게 카테고리를 조금이라도 줄여가면서 데이터 유추가 가능!
```
